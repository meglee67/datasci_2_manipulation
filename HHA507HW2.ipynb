### **1. Data Cleaning with Pandas**

## Identify and handle the missing values in the dataset

```python
# Checking how many values are missing
df.isnull().values.sum()
```

```python
# Replace 'missing' strings with  NaN, so that it shows up as a missing value.
df.replace('missing', np.nan, inplace=True)
df
```

## Removing any duplicate columns and rows

```python
# Identify and remove duplicate rows and columns
df = df.drop_duplicates()
df
# After removing rows you see the number dropped from 105000 rows previously to 100000 rows
```

```python
# Checking for duplicated columns
df.columns.duplicated()
```

```python
# Check for duplicated columns again because I didn't like how the result displayed above
has_duplicate_columns = df.columns.duplicated().any()

if has_duplicate_columns:
    print("There are duplicated columns in the dataset.")
else:
    print("There are no duplicated columns in the dataset.")
```

## Cleaning Column Names

```python
# Removing leading and trailing whitespace
df.columns = df.columns.str.strip()
df
```

```python
# Replace the spaces with underscores for better potential compatibility
df.columns = df.columns.str.replace(' ', '_')
df
```

### 2. Data Transformation

```python
# Creating a new column based on the existing Has_Insurance and Payment_Method

# Create a new column 'payment_risk' based on conditions using np.where
df['Payment_Risk'] = np.where(
    (df['Has_Insurance'] == 'Yes') & (df['Payment_Method'] == 'Insurance'), 'low',
    np.where((df['Has_Insurance'] == 'Yes') & (df['Payment_Method'].isin(['Cash', 'Card'])), 'medium',
    np.where((df['Has_Insurance'] == 'No') & (df['Payment_Method'] == 'Card'), 'high', 'unknown')))

selected_data = ['Has_Insurance', 'Payment_Method', 'payment_risk']
pr = df[selected_data]
pr
```

```python
# Aggregate data using groupby
df.groupby(['Gender']).describe()
```

```python
# Compute summary statistics
stat = df[['Average_Heart_Rate','Average_BP','Patient_Age','Number_of_Tests']]
stat.describe()
```

### 3. Intro to Alternative Libraries

```python
import polars as pl
import pandas as pd
import time
```

```python
# Comparing load times of Polars vs. Pandas
dataset_path = 'healthcare_data_cleaning.csv'

# Start time capture
start_time = time.time()

# Loading dataset using Polars
pandas_df = pd.read_csv(dataset_path)
pandas_time = time.time() - start_time

# Loads dataset using Pandas
polars_df = pl.read_csv(dataset_path)
polars_time = time.time() - start_time

# Print results and load times
print(f"Polars Load Time: {polars_time:.4f} seconds")
print(f"Pandas Load Time: {pandas_time:.4f} seconds")
```

# Observation
* Pandas is just a little bit faster at loading in the dataset compared to Polars.
